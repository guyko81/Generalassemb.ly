One of the best description for variable importance for binary target variable is the so-called Information Value (IV) that is calculated by 
- binnig of the input variable
- calculating the distribution of target (event) and 1-target (non-event) over the bins (sums to 1)
- apply the following equation


```python
# Source: https://github.com/Sundar0989/WOE-and-IV/blob/master/WOE_IV.ipynb
# GNU General Public License v3.0

train_df['target'] = train_df['diagnosis'].apply(lambda x : 1 if x == 'M' else 0)  # Convert to numeric

import pandas.core.algorithms as algos
from pandas import Series
import scipy.stats.stats as stats
import re
import traceback
import string

max_bin = 10
force_bin = 3
eps = 0.01


# define a binning function
def mono_bin(Y, X, n = max_bin):
    
    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X','Y']][df1.X.isnull()]
    notmiss = df1[['X','Y']][df1.X.notnull()]
    r = 0
    while np.abs(r) < 1:
        try:
            d1 = pd.DataFrame({"X": notmiss.X, "Y": notmiss.Y, "Bucket": pd.qcut(notmiss.X, n)})
            d2 = d1.groupby('Bucket', as_index=True)
            r, p = stats.spearmanr(d2.mean().X, d2.mean().Y)
            n = n - 1 
        except Exception as e:
            n = n - 1

    if len(d2) == 1:
        n = force_bin         
        bins = algos.quantile(notmiss.X, np.linspace(0, 1, n))
        if len(np.unique(bins)) == 2:
            bins = np.insert(bins, 0, 1)
            bins[1] = bins[1]-(bins[1]/2)
        d1 = pd.DataFrame({"X": notmiss.X, "Y": notmiss.Y, "Bucket": pd.cut(notmiss.X, np.unique(bins),include_lowest=True)}) 
        d2 = d1.groupby('Bucket', as_index=True)
    
    d3 = pd.DataFrame({},index=[])
    d3["MIN_VALUE"] = d2.min().X
    d3["MEDIAN_VALUE"] = d2.median().X
    d3["MAX_VALUE"] = d2.max().X
    d3["COUNT"] = d2.count().Y
    d3["EVENT"] = d2.sum().Y
    d3["NONEVENT"] = d2.count().Y - d2.sum().Y
    d3=d3.reset_index(drop=True)
    
    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE':np.nan},index=[0])
        d4["MEDIAN_VALUE"] = np.nan
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4,ignore_index=True)
    
    d3["EVENT_RATE"] = d3.EVENT/d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT/d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT/d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT/d3.sum().NONEVENT
    d3["WOE"] = np.log((d3.DIST_EVENT+eps)/(d3.DIST_NON_EVENT+eps))
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT)*np.log((d3.DIST_EVENT+eps)/(d3.DIST_NON_EVENT+eps))
    d3["VAR_NAME"] = "VAR"
    d3 = d3[['VAR_NAME','MIN_VALUE', 'MEDIAN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE', 'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT','DIST_NON_EVENT','WOE', 'IV']]       
    #d3 = d3.replace([np.inf, -np.inf], 0)
    d3 = d3.replace([-np.inf], -5)
    d3 = d3.replace([np.inf], 5)
    d3.IV = d3.IV.sum()
    
    return(d3)

def char_bin(Y, X):
        
    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X','Y']][df1.X.isnull()]
    notmiss = df1[['X','Y']][df1.X.notnull()]    
    df2 = notmiss.groupby('X',as_index=True)
    
    d3 = pd.DataFrame({},index=[])
    d3["COUNT"] = df2.count().Y
    d3["MIN_VALUE"] = df2.sum().Y.index
    d3["MEDIAN_VALUE"] = d3["MIN_VALUE"]
    d3["MAX_VALUE"] = d3["MIN_VALUE"]
    d3["EVENT"] = df2.sum().Y
    d3["NONEVENT"] = df2.count().Y - df2.sum().Y
    
    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE':np.nan},index=[0])
        d4["MEDIAN_VALUE"] = np.nan
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4,ignore_index=True)
    
    d3["EVENT_RATE"] = d3.EVENT/d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT/d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT/d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT/d3.sum().NONEVENT
    d3["WOE"] = np.log((d3.DIST_EVENT+eps)/(d3.DIST_NON_EVENT+eps))
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT)*np.log((d3.DIST_EVENT+eps)/(d3.DIST_NON_EVENT+eps))
    d3["VAR_NAME"] = "VAR"
    d3 = d3[['VAR_NAME','MIN_VALUE', 'MEDIAN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE', 'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT','DIST_NON_EVENT','WOE', 'IV']]      
    #d3 = d3.replace([np.inf, -np.inf], 0)
    d3 = d3.replace([-np.inf], -5)
    d3 = d3.replace([np.inf], 5)
    d3.IV = d3.IV.sum()
    d3 = d3.reset_index(drop=True)
    
    return(d3)

def data_vars(df1, target):
    
    stack = traceback.extract_stack()
    filename, lineno, function_name, code = stack[-2]
    vars_name = re.compile(r'\((.*?)\).*$').search(code).groups()[0]
    final = (re.findall(r"[\w']+", vars_name))[-1]
    
    x = df1.dtypes.index
    count = -1
    
    for i in x:
        if i.upper() not in (final.upper()):
            if np.issubdtype(df1[i], np.number) and len(Series.unique(df1[i])) > 2:
                conv = mono_bin(target, df1[i])
                conv["VAR_NAME"] = i
                count = count + 1
            else:
                conv = char_bin(target, df1[i])
                conv["VAR_NAME"] = i            
                count = count + 1
                
            if count == 0:
                iv_df = conv
            else:
                iv_df = iv_df.append(conv,ignore_index=True)
    
    iv = pd.DataFrame({'IV':iv_df.groupby('VAR_NAME').IV.max()})
    iv = iv.reset_index()
    return(iv_df,iv)


final_iv, IV = data_vars(train_df,train_df['target'])
IV
```

#### Logistic regression

The advantage of using Logistic Regression is that the model gives a nice equation between the input variables and the output. It's easy to manipulate the input and recalculate the prediction.

The disadvantage is that the input variables can correlate heavily which leads to model overfitting. Several techniques present that decreases the risk of overfitting - like L1 and L2 regularisation - however without handling the correlated variables it is highly likely that the coefficients for each input variable won't be in line with the increasing/decreasing probability of the outcome. 

Logistic regression works best with variables that are linear against the WoE value. So the input variable should be changed according to our automatic binning method. 


```python
train_df_logreg = train_df.copy()
features = [col for col in headers if col not in ['ID', 'diagnosis', 'target']]
for col in features:
    bins = [-np.inf]
    bins.extend(final_iv.loc[final_iv['VAR_NAME']==col, 'MIN_VALUE'].values.tolist()[1:])
    bins.append(np.inf)
    labels = final_iv.loc[final_iv['VAR_NAME']==col, 'WOE'].values.tolist()
    train_df_logreg[col] = pd.cut(train_df[col], bins=bins, labels=labels)
train_df_logreg
```

```python
from sklearn.linear_model import LogisticRegression
# I won't use the confusion matrix now
#from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc

train, val, y_train, y_val = train_test_split(train_df_logreg[features], train_df_logreg['target'], test_size=0.2, random_state=42)

logreg = LogisticRegression(penalty='l2', C=1e5) # l2 penalty forces the model to select small values for variables, avoiding overfitting

logreg.fit(train, y_train)
prob = logreg.predict_proba(val)
pred = logreg.predict(val)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()


# Compute micro-average ROC curve and ROC area
fpr["M"], tpr["M"], _ = roc_curve(y_val.ravel(), pred.ravel())
roc_auc["M"] = auc(fpr["M"], tpr["M"])


plt.figure()
lw = 2
plt.plot(fpr["M"], tpr["M"], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc["M"])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
coefficients = pd.DataFrame(logreg.coef_, columns=features)
coefficients = coefficients.T
coefficients.columns=['coefficient']
coefficients = coefficients.join(IV.set_index('VAR_NAME'))
coefficients['important'] = np.abs(coefficients['coefficient']*coefficients['IV'])
coefficients.sort_values(by='important', ascending=False)
```

The most important variables are ['perimeter_sd_error', 'smoothness_sd_error', 'concave_points_sd_error', 'concave_points_worst']

Using the WoE method we can be sure that the negative coefficients are wrong (caused by overfitting).
In this analysis I have left the negative coefficients in the final model, though it is almost sure that dropping correlated variables would increase model performance (I simply don't have the time to iterate the whole process now). 

#### Random Forest

Building a random forest model is a very good bet on model performance. Though it is not optimised on finding the best set of trees (compared to XGBoost or LightGBM methods), but the random set of trees will eventually ensure that the final prediction is not overfitted. 

Other advantage is that we can use the input data table as-is, because the random forest is a non-parametric model.

```python
from sklearn.ensemble import RandomForestClassifier


train, val, y_train, y_val = train_test_split(train_df[features], train_df['target'], test_size=0.2, random_state=42)

# avoiding overfitting: number of trees fairly small, minimum sample in leaf fairly large
rf = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=5, min_samples_split=15, min_samples_leaf=10, 
                            min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, 
                            min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=None, 
                            random_state=None, verbose=0, warm_start=False, class_weight=None)

rf.fit(train, y_train)
prob = rf.predict_proba(val)
pred = rf.predict(val)



# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()


# Compute micro-average ROC curve and ROC area
fpr["M"], tpr["M"], _ = roc_curve(y_val.ravel(), pred.ravel())
roc_auc["M"] = auc(fpr["M"], tpr["M"])


plt.figure()
lw = 2
plt.plot(fpr["M"], tpr["M"], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc["M"])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

The model I first used was Logistic Regression. This model creates an equation that is linearly dependent from each of the input variables. The output value of the equation is then transformed to values between 0 and 1. The larger the outcome the closer the transformed value to 1.

As it is a linear equation we must ensure that the input variables and the target variable has a linear relationship before we fit the model. It is a bit tricky for logistic regression as we are trying to make a model on the inversely transformed target variable (0 becomes very negative value, 1 becomes very positive value). To be able to solve this problem it is common to use the so-called WoE method. Without any technical details let's just approve that this transformation of the input variables will generate new input variables that are optimal for our linear regression (one can check the details in the code). 

An important step is to generate an equation that is not overfitted on the training sample. For avoiding overfit we used L2 regularisation, which forces the model to generate an equation that has low coefficients. Interestingly the low coefficient values indicates low overfitting. 


My second approach was random forest. This model is basically just an average of several (100 in our case) decision tree's prediction. The only trick is that each of the decision trees were built on a random subset of the training data which results in slightly different decision trees. By averaging them we can eliminate the errors caused by each weak model and the result becomes a robust prediction.
